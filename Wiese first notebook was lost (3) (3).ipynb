OK, so we have a graph with very little effort.  Now let's use some scientific conventions and give it a title, as well as label the axes.  For a title, Lets use "Number of Chains Found in a Small Subset of Proteins" and as plotted, the x-axis is "Number of Chains" and the y-axis is "Number of Proteins"  In the following commands, keep the "'" and change Text to the appropriate words.
```python
plt.scatter(chains, counts)
plt.title('Number of Chains Found in a Small Subset of Proteins')
plt.xlabel('Number of Chains')
plt.ylabel('Number of Proteins')
```

There are a number of types of graphs that can be produced:
plt.scatter(x,y): yields scatter plot with just markers
    different markers can be used: try s, o, ^
    lines can be used to connect points if appropriate: try -, --, and . (not all work)
plt.plot(x,y): yields line plot, markers optional
plt.bar(x,y): yields bar plot
plt.stem(x,y): yields stem plot (like scatter plot with lines to x-axis
plt.boxplot(x,y): yields box plot
plt.hist(nums): yields histogram plot showing distribution of values in dataset
plt.pie(nums): yields a pie plot showing relative ratios
